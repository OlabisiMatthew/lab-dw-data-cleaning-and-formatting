## Exercise 1: Cleaning Column Names

To ensure consistency and ease of use, standardize the column names of the dataframe. Start by taking a first look at the dataframe and identifying any column names that need to be modified. Use appropriate naming conventions and make sure that column names are descriptive and informative.

*Hint*:
- *Column names should be in lower case*
- *White spaces in column names should be replaced by `_`*
- *`st` could be replaced for `state`*

```python
import numpy as np
import pandas as pd
```

```python
#reading the data
customer_data =pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv")
customer_data.head()
```

```python
customer_data.columns
```

```python
#changing the column names to lowercase

column_names=customer_data.columns
cols = []
for colname in customer_data.columns:
    cols.append(colname.lower())

customer_data.columns = cols

display(customer_data.head())
```

```python


customer_data = customer_data.rename(columns={'st':'state', 
                              'customer lifetime value': 'customer_lifetime_value', 
                              'monthly premium auto': 'monthly_premium_auto', 
                              'number of open complaints': 'number_of_open_complaints',
                               'policy type': 'policy_type', 
                              'vehicle class': 'vehicle_class', 
                              'total claim amount': 'total_claim_amount'
                             })

display(customer_data.head())
```

## Exercise 2: Cleaning invalid Values

The dataset contains columns with inconsistent and incorrect values that could affect the accuracy of our analysis. Therefore, we need to clean these columns to ensure that they only contain valid data.

Note that this exercise will focus only on cleaning inconsistent values and will not involve handling null values (NaN or None).

*Hint*:
- *Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".*
- *State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*
- *In education, "Bachelors" could be replaced by "Bachelor"*
- *In Customer Lifetime Value, delete the `%` character*
- *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*

```python
# Cleaning the invalid values in the gender column.

df2 = customer_data.copy()

df2['gender'].replace('Femal', 'F', inplace=True)
df2['gender'].replace('female', 'F', inplace=True)
df2['gender'].replace('Male', 'M', inplace=True)

df2['gender'].unique()
```

```python
# Replacing state abbreviations with their full names.
df2['state'].replace({'AZ':'Arizona', 'Cali': 'California', 'WA': 'Washington' }, inplace=True)

df2['state'].unique()
```

```python
#Replacing Bachelors with bachelor
df2['education'].replace('Bachelors', 'Bachelor', inplace=True)

df2['education'].unique()
```

```python
# Deleting the '%' character from the values of Customer Lifetime Value
df2['customer_lifetime_value'] = df2['customer_lifetime_value'].str.replace('%', '').astype(float)
df2['customer_lifetime_value'].unique()
```

```python
df2['vehicle_class'].replace({'Sports Car':'Luxury', 'Luxury SUV': 'Luxury', 'Luxury Car': 'Luxury' }, inplace=True)

df2['vehicle_class'].unique()
```

## Exercise 3: Formatting data types

The data types of many columns in the dataset appear to be incorrect. This could impact the accuracy of our analysis. To ensure accurate analysis, we need to correct the data types of these columns. Please update the data types of the columns as appropriate.

It is important to note that this exercise does not involve handling null values (NaN or None).

pd.to_numeric*Hint*:
- *Customer lifetime value should be numeric*
- *Number of open complaints has an incorrect format. Look at the different values it takes with `unique()` and take the middle value. As an example, 1/5/00 should be 5. Number of open complaints is a string - remember you can use `split()` to deal with it and take the number you need. Finally, since it should be numeric, cast the column to be in its proper type.*

```python
# Changing the data types of the two variables from string to numeric

df2['customer_lifetime_value'] =pd.to_numeric((df2['customer_lifetime_value']), errors='coerce')
 
print(df2.info()) 
```

```python
# Splitting the 'number_of_open_complaints' using '/' as the delimiter
df2['number_of_open_complaints'] = df2['number_of_open_complaints'].str.split('/').str[1]

df2['number_of_open_complaints'] =(df2['number_of_open_complaints']).apply (pd.to_numeric)

print(df2.info()) 
```

Identify any columns with null or missing values. Identify how many null values each column has. You can use the `isnull()` function in pandas to find columns with null values.

Decide on a strategy for handling the null values. There are several options, including:

```python
# Checking for missing values in each column
df2.isna().sum()
```

```python
# Instead of dropping rows or coloumns with missing values, i would like to fill in the mean of values and mode of classes for numerical and categorical variables respectively. This strategy was chosen so as to to drop the large volume of observations because of missing values.
#FOR NUMERICAL COLUMNS
mean_customer_lifetime_value = df2['customer_lifetime_value'].mean()
df2['customer_lifetime_value'] = df2['customer_lifetime_value'].fillna(mean_customer_lifetime_value)

mean_income = df2['income'].mean()
df2['income'] = df2['income'].fillna(mean_income)

mean_monthly_premium_auto = df2['monthly_premium_auto'].mean()
#print("The mean value of the column {} is {}".format('monthly_premium_auto',mean_monthly_premium_auto))
df2['monthly_premium_auto'] = df2['monthly_premium_auto'].fillna(mean_monthly_premium_auto)
                    
mean_number_of_open_complaints = df2['number_of_open_complaints'].mean()
df2['number_of_open_complaints'] = df2['number_of_open_complaints'].fillna(mean_number_of_open_complaints)

mean_total_claim_amount = df2['total_claim_amount'].mean()
df2['total_claim_amount'] = df2['total_claim_amount'].fillna(mean_total_claim_amount)


#FOR CATEGORICAL VARIABLES
mode_state = df2['state'].mode()
df2['state'] = df2['state'].fillna(mode_state)[0]

mode_gender = df2['gender'].mode()
df2['gender'] = df2['gender'].fillna(mode_gender)[0]

mode_education = df2['education'].mode()
df2['education'] = df2['education'].fillna(mode_education)[0]

mode_policy_type = df2['policy_type'].mode()
df2['policy_type'] = df2['policy_type'].fillna(mode_policy_type)[0]

mode_vehicle_class = df2['vehicle_class'].mode()
df2['vehicle_class'] = df2['vehicle_class'].fillna(mode_vehicle_class)[0]

df2.isna().sum()
```
